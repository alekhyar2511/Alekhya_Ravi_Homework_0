```python
keys = data_frame_copy.columns
keys
```

```python
dictionary = {};
for key_name in keys:
    dictionary[key_name] = [];
```

```python
dictionary
```

```python
key_dictionary = {};
for key_name in keys:
    for index, row_data in data_frame_copy.iterrows():
        key_dictionary[index] = row_data[key_name];
    dictionary[key_name] = key_dictionary;
    key_dictionary = {};
dictionary
```

```python
new_data_frame = pd.DataFrame(dictionary)
```
